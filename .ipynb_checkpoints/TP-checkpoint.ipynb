# V) Nettoyage des données

### 1) Modification des noms des colonnes

```python
# Verifions d'abort comment les noms de colonnes sont écrit :
data.columns
```

#### a) Effacement des éspaces vide se trouvant devant un nom de la colonne

```python
# Pour effacer ces éspaces vides,on va créer une boucle qui va parcourir chaque nom et s'il trouve un
# éspace vide devant ce nom,il va l'éffacer
for col in data.columns : # Dans les colonnes du Dataframe
    if col[0] == " " : # S'il y'a dans le nom de la colonne un éspace vide qui commence
        data.rename(columns={col:col.replace(col[0],"")}, inplace=True) # Alors,on va l'énlever
data.columns # On verifie et on voit qu'on a pu effacer l'éspace vide de devant pour certains colonnes comme la colonne "ID"
```

#### b) Effacement des "?" (point d'interrogation) se trouvant dans certains noms des colonne

```python
# Pour effacer les "?",on va créer une boucle qui va parcourir chaque nom et s'il trouve un
# point virgule dans le nom,il va l'éffacer
for col in data.columns : # Dans les colonnes du Dataframe
    if "?" in col : # S'il y'a dans le nom de la colonne un poit d'iterrogatio
        data.rename(columns={col:col.replace("?","")}, inplace=True) # Alors,on va l'énlever
data.columns # On verifie et on voit qu'on a pu effacer les points virgules pour les colonnes comme la colonne "Literate?"
```

#### c) Effacement des "," (virgule) se trouvant dans certains noms des colonne

```python
# Pour effacer les ",",on va créer une boucle qui va parcourir chaque nom et s'il trouve une
# virgule dans le nom,il va l'éffacer
for col in data.columns : # Dans les colonnes du Dataframe
    if "," in col : # S'il y'a dans le nom de la colonne une virgule
        data.rename(columns={col:col.replace(",","")}, inplace=True) # Alors,on va l'énlever
data.columns # On verifie et on voit qu'on a pu effacer les points virgules pour les colonnes comme la colonne "If sorghum, type"
```

#### d) Remplacement des éspaces par des "_" dans les noms des colonnes

```python
# Pour remplacer les noms des éspaces par des "_", on fait:
data.columns=data.columns.str.replace(" ","_") # On remplace les éspaces par "_"
data.columns # Puis on affiche pour voir les changements
```

### 2) Suppression des colonnes ayant des NaN > 90 ou une seule modalité

#### a) Recherche des colonnes plus de 90% de NaN

```python
# Pour réaliser cette tache on va créer une bloucle qui va parcourir chaque colonne puis calculer 
#le pourcentage de ces données maquantes
misCol=[] # On déclare une liste qui va contenir les colonnes ayant plus de 90% de NaN
print("Voici les colonnes ayant plus de 90% de NaN :")
for col in data.columns:
    pource=round((data[col].isnull().sum()/data.shape[0])*100,2) # On calcule le pourcentage d'une colonne puis on l'arrondi
    if pource >= 90 : # Si le pourcentage est superieur à 90
        misCol.append(col) # Alors,on va enregistrer la colonne dans la liste
        print("=>La colonne '{}', il a {}% de données maquantes".format(col, pource)) # Puis on affiche la colonne et son pourcentage de NaN
```

```python
# Après avoir trouver ces colonnes, on va alors les supprimées en faisant:
data.drop(misCol, axis=1, inplace=True)
# Puis on va verifier pour voir si la colonne ont été supprimer avec succès
for mis in misCol:
    if mis in data.columns : # Si la colonne est encore dans le dataframe
        print("True") # Afficher "True"
    else: # Si non
        print("False") # Afficher "False"
```

#### b) Recherche des colonnes ayant une seul modalité

```python
# Pour réaliser cette tache on va créer une bloucle qui va parcourir chaque colonne puis chercher s'il 
#y'a une seul modalité
modCol=[] # On déclare une liste qui va contenir les colonnes ayant une seul modalité
print("Les colonnes ayant une seul modalité sont :")
for col in data.columns:
    nbMod=len(list(data[col].unique())) # On va compter le nombre de modalité
    if nbMod == 1 : # Si la colonne a une seul modalité
        modCol.append(col) # Alors,on va enregistrer la colonne dans la liste
        print("=>La colonne '{}'".format(col)) # Puis on affiche la colonne et son pourcentage de NaN
```

```python
# Après avoir trouver ces colonnes, on va alors les supprimées en faisant:
data.drop(modCol, axis=1, inplace=True)
# Puis on va verifier pour voir si les colonnes ont été supprimer avec succès
for mis in modCol:
    if mis in data.columns : # Si la colonne est encore dans le dataframe
        print("True") # Afficher "True"
    else: # Si non
        print("False") # Afficher "False"
```

```python
# Créons alors 3 fonctions qui vont nous servir dans cette taches

# Cette fonction va nous servir de transformer tous les données de la colonne en MAJUSCULE
def majuscule(nom_de_la_colonne):
    # On créer un dictionnaire qui va contenir les données de la colonnes et leur numero de ligne
    diction=dict(data[nom_de_la_colonne])
    # On va alors créer une boucle qui va tous transformer en majuscule
    for index, val in diction.items():
        diction[index]=val.upper()
    # Après les modifications on va affecter les données modifier dans le dataframe
    data[nom_de_la_colonne]=diction.values()
    
    
# Cette fonction va nous servir de transformer tous les données de la colonne en minuscule
def minuscule(nom_de_la_colonne):
    # On créer un dictionnaire qui va contenir les données de la colonnes et leur numero de ligne
    diction=dict(data[nom_de_la_colonne])
    # On va alors créer une boucle qui va tous transformer en minuscule
    for index, val in diction.items():
        diction[index]=val.lower()
    # Après les modifications on va affecter les données modifier dans le dataframe
    data[nom_de_la_colonne]=diction.values()
    
    
# Cette fonction va nous servir de transformer tous les données de la colonne en mettant la
# premier lettre en majuscule et le reste en minuscule
def PremierMajuscule(nom_de_la_colonne):
    # On créer un dictionnaire qui va contenir les données de la colonnes et leur numero de ligne
    diction=dict(data[nom_de_la_colonne])
    # On va alors créer une boucle qui va tous transformer la première lettre en majuscule et le reste en minuscule
    for index, val in diction.items():
        diction[index]=val.capitalize()
    # Après les modifications on va affecter les données modifier dans le dataframe
    data[nom_de_la_colonne]=diction.values()
    
```
